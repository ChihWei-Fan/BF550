```python
import numpy as np
import statistics as stat
from scipy import stats
np.set_printoptions(precision=3)
np.random.seed(20221110)
import matplotlib.pyplot as plt
sns.set()
```

```python
# Make thousands of runs to get better statistics.
Nruns= 1000
# The max steps for each run is 1000
Nsteps= 1000

#give three empty list to store the final position at t=10,t=100, t=1000 
finalpos_10=[]
finalpos_100=[]
finalpos_1000=[]

#use for loop to loop over several walks

for j in range(Nruns):
    #create an array with zeros, to keep track of each step later
    pos=np.zeros(Nsteps)
    #create three variables to store the postion at specific time in each run
    pos_10 = 0
    pos_100 = 0
    pos_1000 = 0
    #Loop over the steps of a single walk
    for i in range(Nsteps):
        # get a random floating point number btw 0.0 and 1.0.
        prob = np.random.random()
        if prob <1/3:  #turn left current postion -1
            pos[i]=pos[i-1]-1  
        elif 1/3 < prob <2/3: #turn right current postion +1
            pos[i]=pos[i-1]+1 
        else:
            pos[i]=pos[i-1] #not move 
            
        # when the t=10 store the postion
        if i == 9:
            pos_10 = pos[i]
            
        # when the t=100 store the postion   
        if i == 99:
            pos_100 = pos[i]
            
        # when the t=1000 store the postion
        if i == 999:
            pos_1000 = pos[i]
    #append the postion at different timing that was stored 
    #forming lists that store the final postion at different timing at each run 
    finalpos_10.append(pos_10)
    finalpos_100.append(pos_100)
    finalpos_1000.append(pos_1000)
    
#sorted the data, so they could be drawn as curves
finalpos_10 = sorted(finalpos_10)
finalpos_100 = sorted(finalpos_100)
finalpos_1000 = sorted(finalpos_1000)

    
#calculte the mean and standard deviation in different t.

mean10=np.mean(finalpos_10)
std10=np.std(finalpos_10)
print("t=10 mean:",mean10,"t=10 std:",std10)

mean100=np.mean(finalpos_100)
std100=np.std(finalpos_100)
print("t=100 mean:",mean100,"t=100 std:",std100)

mean1000=np.mean(finalpos_1000)
std1000=np.std(finalpos_1000)
print("t=1000 mean:",mean1000,"t=1000 std:",std1000)

#print the plot
plt.figure(figsize=(8,5))
plt.plot(finalpos_10,stats.norm.pdf(finalpos_10,loc=mean10,scale=std10),label='t=10 P.D.F.')
plt.plot(finalpos_100,stats.norm.pdf(finalpos_100,loc=mean100,scale=std100),label='t=100 P.D.F.')
plt.plot(finalpos_1000,stats.norm.pdf(finalpos_1000,loc=mean1000,scale=std1000),label='t=1000 P.D.F.')
plt.xlim([-150, 150])
plt.title("Probability Density Function",fontsize= 15)
plt.xlabel("Final_Position",fontsize= 15)
plt.ylabel("Probability_Density",fontsize= 15)
plt.legend()
plt.show()
```
